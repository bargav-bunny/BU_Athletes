```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/dataframe_with_extension.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# Selecting All polling features, along wiht Team Name and Cumulative BU GPA

pd.concat([df[['Team Name', 'Last Cumulative BU GPA']], df.iloc[:, 66:]], axis = 1).groupby('Team Name').agg('mean').to_csv('gpa_and_polling.csv', index = False)
```

```python
df = pd.read_csv('gpa_and_polling.csv')
df.head()
```

```python
# Calculating the correlation of 'Last Cumulative BU GPA' with every other column

correlations = df.corr()['Last Cumulative BU GPA'].sort_values().iloc[:-1]
```

```python
correlations
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=correlations.values, y=correlations.index)
plt.title('Correlation of Each Polling Feature with Averge Cumulative BU GPA of that team')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.savefig('corr_polling_gpa.png')
plt.show()
```
