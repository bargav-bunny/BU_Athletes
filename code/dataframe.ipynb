Importing the data into pandas df and getting rid of some columns that are unnecessary.

```python
import pandas as pd

df = pd.read_csv("../data/BU Athletics Data 22-23.csv")  
del df['Student']
del df['US Region']

```

Getting rid of all of the columns that have equal values for each row. This is mostly just the last few columns GPA values that are all 0.0 for every athlete.

```python
nunique = df.nunique()
cols_to_drop = nunique[nunique == 1].index
for col in cols_to_drop:
    del df[col]
df
```

Renaming some columns to less complicated descriptions.

```python
df = df.rename(columns={'English Language Primary Language of Country': 'Language', 'Athletic Code 1': 
                        'Athletic Code', 'Athletic Code Description 1': 'Team Name', 'First Major (HEGIS) Description': 
                        'First Major Name', 'Last Major (HEGIS) Description': 'Last Major Name'})
df
```

Filling in some 'Nan' values.

```python
df[['Country']] = df[['Country']].fillna('USA')
df[['Language']] = df[['Language']].fillna('English')
df
```
