```python
import numpy as np  
import matplotlib.pyplot as plt  
  
X = ['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9','Q10','Q11','Q12','Q13','Q14','Q15','Q16','Q17'] 
  
X_axis = np.arange(len(X)) 

plt.figure(figsize=(12, 6))

plt.bar(X_axis - 0.2, g, 0.4, label = 'W') 
plt.bar(X_axis + 0.2, b, 0.4, label = 'M') 
  
plt.xticks(X_axis, X) 
plt.xlabel("Questions") 
plt.ylabel("Polling Value") 
plt.title("Average Polling Question Results for Mens versus Womens Sports") 
plt.legend() 
plt.show() 
```

```python
difference = abs(np.array(g) - np.array(b))

# Find the indices of X values with the most difference
max_diff_indices = np.argsort(difference)[-5:]  # Adjust to get top 'n' differences
max_diff_X_values = [X[i] for i in max_diff_indices]

print(f"X values with the most difference: {max_diff_X_values}")

# Find the top 3 X values where 'g' is most greater than 'b'
g_greater_indices = [i for i in range(len(g)) if g[i] > b[i]]
if len(g_greater_indices) > 0:
    sorted_g_greater_indices = sorted(g_greater_indices, key=lambda x: g[x] - b[x], reverse=True)[:3]
    max_g_greater_X_values = [X[i] for i in sorted_g_greater_indices]
    print(f"Top 3 X values where 'g' is most greater than 'b': {max_g_greater_X_values}")
else:
    print("There are no X values where 'g' is greater than 'b'.")

# Find the top 3 X values where 'b' is most greater than 'g'
b_greater_indices = [i for i in range(len(b)) if b[i] > g[i]]
if len(b_greater_indices) > 0:
    sorted_b_greater_indices = sorted(b_greater_indices, key=lambda x: b[x] - g[x], reverse=True)[:3]
    max_b_greater_X_values = [X[i] for i in sorted_b_greater_indices]
    print(f"Top 3 X values where 'b' is most greater than 'g': {max_b_greater_X_values}")
else:
    print("There are no X values where 'b' is greater than 'g'.")
```
