```python
import numpy as np
major_switch = []
transfer = []
gpa_diff = []
tested = []
for i, row in df.iterrows():
    if row['First Major Name'] == row['Last Major Name']:
        major_switch.append(0)
    else:
        major_switch.append(1)

    if row['FCLASS'] != "TRANSFERS":
        transfer.append(0)
    else:
        transfer.append(1)

    gpa_diff.append(row['Last Cumulative BU GPA'] - row['Cumulative GPA-1st Sem'])

    if row['SATCFORAI'] == None and row['ACTUSE'] == None:
        tested.append(0)
    else:
        tested.append(1)

df['SwitchedMajor'] = np.array(major_switch)
df['transfer'] = np.array(transfer)
df['Last Cumulative and First GPA diff'] = np.array(gpa_diff)
df['Standardized Test'] = np.array(tested)
```

```python
import copy
import matplotlib.pyplot as plt
from statistics import median
import numpy as np

df_copy = copy.deepcopy(df)
df_copy = df_copy.dropna(subset = ['Cumulative GPA-1st Sem', 'Cumulative GPA-2nd Sem', 'Cumulative GPA-3rd Sem', 'Cumulative GPA-4th Sem', 'Cumulative GPA-5th Sem', 'Cumulative GPA-6th Sem', 'Cumulative GPA-7th Sem', 'Last Cumulative BU GPA']) 

team_dict = {}
for i, row in df_copy.iterrows():
    if row['Team Name'] not in team_dict:
        team_dict[row['Team Name']] = []
    team_dict[row['Team Name']].append([row['Cumulative GPA-1st Sem'], row['Cumulative GPA-2nd Sem'], row['Cumulative GPA-3rd Sem'], row['Cumulative GPA-4th Sem'], row['Cumulative GPA-5th Sem'], row['Cumulative GPA-6th Sem'], row['Cumulative GPA-7th Sem'], row['Last Cumulative BU GPA']])

team_dict_final = {}
team_lines = []
for key in team_dict.keys():
    mat = team_dict[key]
    arr = np.mean(np.array(mat), axis=0)
    team_dict_final[key] = arr
    team_lines.append(arr)
print(team_dict_final)
print(team_lines)

for key in team_dict_final.keys():
    plt.plot(team_dict_final[key], label=str(key))

plt.title("Athletes GPAs by semester")
plt.xlabel("Semester")
plt.ylabel("GPA")
plt.legend(loc="lower right", prop={'size': 4})
plt.show()
```

We now have two new data frames, one that groups GPA by the team and major as well as another data frame that is just for athletes who changed major. New features include the athletes last cumulative GPA - first GPA, a feature indicating if the student took standardized tests, and a feature indicating if the student is a transfer student.

```python
df = pd.read_csv('../data/GPA_major_gender_grouped_count.csv')
counts = df.groupby(['Last Major Name', 'Team Gender']).size().unstack()

# Filter majors where both male and female GPAs are available
valid_majors = counts[(counts['M'].notnull()) & (counts['W'].notnull())].index

# Filter the original DataFrame based on valid majors
filtered_df = df[df['Last Major Name'].isin(valid_majors)]

# Pivot the filtered DataFrame to have genders as columns and majors as index
pivot_df = filtered_df.pivot(index='Last Major Name', columns='Team Gender', values='Last Cumulative BU GPA')

top_10_men = pivot_df['M'].nlargest(6).index.tolist()
top_10_women = pivot_df['W'].nlargest(6).index.tolist()

print("Top 10 Most Popular Men's Majors:")
print(top_10_men)
print("\nTop 10 Most Popular Women's Majors:")
print(top_10_women)

# Calculate majors that differ the most by percentage (Top 10)
difference_percentage = ((pivot_df['W'] - pivot_df['M']) / pivot_df['M']) * 100
top_10_diff = difference_percentage.abs().nlargest(5)

print("\nTop 10 Majors that Differ the Most by Percentage:")
print(top_10_diff)

# Plotting
pivot_df.plot(kind='bar', figsize=(10, 6))
plt.title('Count by Major for Men and Women')
plt.xlabel('Major')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.grid(axis='y')  # Show gridlines on y-axis

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('../data/GPA_major_gender_grouped.csv')
counts = df.groupby(['Last Major Name', 'Team Gender']).size().unstack()

# Filter majors where both male and female GPAs are available
valid_majors = counts[(counts['M'].notnull()) & (counts['W'].notnull())].index

# Filter the original DataFrame based on valid majors
filtered_df = df[df['Last Major Name'].isin(valid_majors)]

# Pivot the filtered DataFrame to have genders as columns and majors as index
pivot_df = filtered_df.pivot(index='Last Major Name', columns='Team Gender', values='Last Cumulative BU GPA')

# Calculate the count of times male GPA is higher than female and vice versa
count_male_higher = (pivot_df['M'] > pivot_df['W']).sum()
count_female_higher = (pivot_df['W'] > pivot_df['M']).sum()
print(f"Count of times male GPA is higher than female: {count_male_higher}")
print(f"Count of times female GPA is higher than male: {count_female_higher}")


# Plotting
pivot_df.plot(kind='bar', figsize=(10, 6))
plt.title('GPA by Major for Men and Women')
plt.xlabel('Major')
plt.ylabel('GPA')
plt.legend(title='Gender')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.grid(axis='y')  # Show gridlines on y-axis

plt.tight_layout()
plt.show()
```
